```
# Plotting the distribution of variables
sns.displot(data=twitchdata, x="Watch time(Minutes)")
sns.displot(data=twitchdata, x="Stream time(minutes)")
sns.displot(data=twitchdata, x="Peak viewers")
sns.displot(data=twitchdata, x="Average viewers")
sns.displot(data=twitchdata, x="Followers")
sns.displot(data=twitchdata, x="Followers gained")
sns.displot(data=twitchdata, x="Views gained") 

```

```
# Plotting the correlation among variables
sns.heatmap(twitchdata.corr(), annot=True, cmap="coolwarm")

```

```
# Boxplot to see the outliers and median distribution
sns.boxplot(data=twitchdata, x="Watch time(Minutes)")

```

```
# Violinplot to see the density of distribution
sns.violinplot(data=twitchdata, x="Stream time(minutes)") 

```

```
# Pairplot to see the distribution and relationship among variables
sns.pairplot(twitchdata)  

```

```
# Countplot to see the frequency of categorical variables
sns.countplot(data=twitchdata, x="Partnered")
sns.countplot(data=twitchdata, x="Mature")
sns.countplot(data=twitchdata, x="Language") 

```

```
# Barplot to see the average value of variables
sns.barplot(data=twitchdata, y="Peak viewers", x="Language") 

```
